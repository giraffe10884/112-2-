```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# CSV
df_2023 = pd.read_csv('2023_world_pool_rankings.csv')
df_2024 = pd.read_csv('2024_world_pool_rankings.csv')

# 1. 2024202320
plt.figure(figsize=(12, 6))
plt.subplot(121)
df_2023['Nationality'].value_counts().plot(kind='bar', title='nationality distribution of the top 20 players in 2023.')
plt.subplot(122)
df_2024['Nationality'].value_counts().plot(kind='bar', title='nationality distribution of the top 20 players in 2024.')
print('Q1)')
plt.show()

# 2. 202320242020
common_players = pd.merge(df_2023, df_2024, on='Player', how='inner', suffixes=('_2023', '_2024'))
print('Q2)')
print('Players who were in the top 20 in both years', common_players['Player'].tolist())
print()

# 3. 20232024
plt.figure(figsize=(10, 6))
for player in common_players['Player']:
    plt.plot([2023, 2024], common_players.loc[common_players['Player'] == player, ['Ranking_2023', 'Ranking_2024']].values[0], label=player)
plt.gca().invert_yaxis()
plt.xticks([2023, 2024])  # x
plt.yticks(np.arange(1, 21))  # y
plt.xlabel('year')
plt.ylabel('ranking')
plt.title('The ranking changes of the players.')
plt.legend()
print('Q3)')
plt.show()
```
